Make a 3D plot showing no direct relationship between S, N, and E, without B.

```python
import numpy as np
import pandas as pd

from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline
```

```python
# Import the data
data = pd.read_csv('data_statevariables.csv')
```

# 3D plots

Show that with S,N,E alone we have a random scatter plot

```python
# Get maxes to make correct color range
emin = np.min(np.log10(data['E']))
emax = np.max(np.log10(data['E']))

# Set up figure
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for s,n,e in zip(data['S'],np.log10(data['N']),np.log10(data['E'])):
    ax.plot([s,s],[n,n],[e,0],'o-',markevery=2,c=cm.winter((e-emin)/(emax-emin)),linewidth=2,alpha=0.7)
ax.set_xlabel('S')
ax.set_ylabel('log10(N)')
ax.set_zlabel('log10(E)')
ax.view_init(25,320)
```

```python
from IPython.display import HTML
from matplotlib import animation
```

```python
# Make an animated version

# Get maxes to make correct color range
emin = np.min(np.log10(data['E']))
emax = np.max(np.log10(data['E']))

# Set up figure
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for s,n,e in zip(np.log10(data['S']),np.log10(data['N']),np.log10(data['E'])):
    ax.plot([s,s],[n,n],[e,0],'o-',markevery=2,c=cm.winter((e-emin)/(emax-emin)),linewidth=2,alpha=0.7)
ax.set_xlabel('log10(S)')
ax.set_ylabel('log10(N)')
ax.set_zlabel('log10(E)')

# rotate the axes and update
def animate(frame):
    ax.view_init(30, frame)
    plt.pause(.001)
    return fig

anim = animation.FuncAnimation(fig, animate, frames=range(30,330), interval=50)
HTML(anim.to_html5_video())
```
